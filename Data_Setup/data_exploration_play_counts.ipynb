# Age Buckets Detailed Analysis

```python
age_data = plays_data[['age','userId','artistId','plays']]
```

```python
age_data.head()
```

```python
age_data['age'].value_counts()
```

```python
age_data[age_data['age']<150].sort_values(by='age')
```

```python
import numpy as np
```

```python
age_data = age_data[np.isfinite(age_data['age'])]
```

```python
def create_bucket(age):
    if age>10 and age<18:
        return 'b1'
    elif age>=18 and age<23:
        return 'b2'
    elif age>=23 and age<28:
        return 'b3'
    elif age>=28 and age<35:
        return 'b4'
    elif age>=35 and age<150:
        return 'b5'
```

```python
age_data['bucket'] = age_data['age'].apply(lambda x: create_bucket(x))
```

```python
age_data.head()
```

```python
age_data.describe()
```

## No of user in each age buckets

```python
age_data[['bucket','userId']].groupby('bucket').nunique()
```

```python
age_aggr = age_data[['bucket','artistId','plays']].groupby(['bucket','artistId'],as_index=False).sum()
```

```python
age_aggr.head()
```

## Artist choice for different age group

```python
age_final = age_aggr[age_aggr['bucket']=='b2'].sort_values(by='plays',ascending=False)
```

```python
age_final
```
